# Retail Forecasting Project

### Data Preparation

```python
import pandas as pd
import datetime as dt
```

```python
retailDF = pd.read_csv('ForecastingCaseStudy.csv')
retailDF
```

```python
retailDF.rename(columns = {'V_DAY': 'Valentines Day', 'EASTER': 'Easter', 'CHRISTMAS': 'Christmas'}, inplace = True)
retailDF
```

```python
yearCollector = []
weekCollector = []
for date in retailDF['date']:
    yearCollector.append(dt.date(int(date.split('/')[2]), 
                                     int(date.split('/')[0]), 
                                     int(date.split('/')[1])).isocalendar().year)
    weekCollector.append(dt.date(int(date.split('/')[2]), 
                                     int(date.split('/')[0]), 
                                     int(date.split('/')[1])).isocalendar().week)
```

```python
retailDF['Year'] = yearCollector
retailDF['Week'] = weekCollector
retailDF
```

```python
retailDF['Price Discount (%)'] = retailDF['Price Discount (%)'].map(lambda x: x.rstrip('%'))
retailDF = retailDF.drop(columns = ['date'])
retailDF
```
